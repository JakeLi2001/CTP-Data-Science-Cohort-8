```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='darkgrid', palette='colorblind')
```

# Question 1

```python
df = pd.read_excel('Data Visualization Exercise Data.xlsx', sheet_name=0)
df.head()
```

```python
xticklabels = ['0']
for x in range(10000, 250000, 10000):
    xticklabels.append(str(x)[:-3]+'k')
```

```python
fig, ax = plt.subplots(figsize=(16, 5))

sns.histplot(x='base_salary', data=df, bins=26, ax=ax).set(title='Base Salary Distribution', xlim=(0, 250000), xticks=range(0, 250000, 10000), xticklabels=xticklabels)
plt.show()
print('Most people have a base salary around 40K to 90k.')
```

```python
df_2 = df.groupby('agency_name')['base_salary'].mean().sort_values(ascending=False).reset_index()
df_2.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.barplot(y='agency_name', x='base_salary', data=df_2, orient='h', order=df_2['agency_name'][0:10], ax=ax).set(title='Average Base Salary per Agency', xticks=range(0, 250000, 25000))
plt.show()
print('Based on the bar plot, I would want to work for the NYC Fire Pension Fund agency.')
```

# Question 4

```python
df4 = pd.read_excel('Data Visualization Exercise Data.xlsx', sheet_name=4)
df4.head()
```

```python
df4_2 = pd.melt(df4, id_vars=['month_number'], value_vars=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'])
df4_2.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

sns.lineplot(x='month_number', y='value', data=df4_2, hue='variable', palette='Set1', ax=ax).set(title='Item Sales Over The Last 12 Months',
                                                                                                 ylabel='sales',
                                                                                                 xlim=(0,12))
plt.show()
print('Based on the line chart, I would recommend to push for more bathing soaps and stop selling moisturizers.')
```

```python
df4['total_sales'] = df4['facecream'] + df4['facewash'] + df4['toothpaste'] + df4['bathingsoap'] + df4['shampoo'] + df4['moisturizer']
df4.head()
```

```python
items = df4.columns[1:7]

for item in items:
    df4[item + '_pct'] = round(df4[item] / df4['total_sales'], 2)
    
df4.head()
```

```python
df4_3 = pd.melt(df4, id_vars=['month_number'], value_vars=['facecream_pct', 'facewash_pct', 'toothpaste_pct', 'bathingsoap_pct', 'shampoo_pct', 'moisturizer_pct'])
df4_3.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))

sns.barplot(x='month_number', y='value', data=df4_3, hue='variable', palette='Set1', ax=ax).set(title='Item Sales in % of Total Sales Over The Last 12 Months',
                                                                                                ylabel='% of total sales')
plt.show()
```
